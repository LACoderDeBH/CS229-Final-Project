```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
"""
INFO:
101,332 Players with up to 20 features each (exluding year, including team)
if metric not reported for player, set to 0.0 by default
"""
batting_data_path = 'bsb_ref.csv'

df = pd.read_csv(batting_data_path)
df = df.fillna(0)  # fixing this could really help with results potentially
```

```python
def find_peers(player_data, n=6):
    """
    Ideally:
        -Drop players who have played fewer than n years.
        -Drop years not in [1, n].
        -Normalize each metric for each year i in n.
        -Compute L2 norm between player and all others.
        -Return new dataframe containing player's peers in range [1, n]
    Currently:
        -Drop players who have played fewer than n years.
        -Drop years not in [1, n].
    Hopefully with a sufficiently large n, the players will all be similar enough.
    """
    player_data = player_data.groupby('name_common').filter(lambda years: len(years) >= n)
    
    players = set(player_data['name_common'].values)
    for p in players:
        player_data[player_data['name_common'] == p] \
            = player_data[player_data['name_common'] == p].head(n=n)
    player_data = player_data.dropna()   # sus af
    return player_data

    
peers = find_peers(pd.DataFrame.copy(df))
```
