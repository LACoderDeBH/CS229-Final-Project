```python
import requests
import re
from bs4 import BeautifulSoup

def iterate_player_list():
    all_player_shtmls = set()
    
    #might be (<a href="\/players\/a\/)(\w+)(.shtml) not sure if esc char needed
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    base = 'https://www.baseball-reference.com/players/'
    for letter in alphabet:
        regex = '(<a href="/players/' + letter + '/)(\w+)(.shtml)'
        resp = requests.get(base + letter)
        soup = BeautifulSoup(resp.content, 'html.parser')
        
        #Parse on hyperlinks
        player_links = str(soup.find_all('', {'href': re.compile(r'/players/' + letter + '/')})) 
        #Hashset to remove dups
        player_shtmls = set(re.findall(regex, player_links)) 
        
        for player_shtml in player_shtmls:
            formatted_link = base + letter + '/' + player_shtml[1] + player_shtml[2]
            all_player_shtmls.add(formatted_link)

    return all_player_shtmls
```

```python
players_to_query = iterate_player_list()

players_to_query
```

```python
import scrapy
from scrapy.crawler import CrawlerProcess

class MarilynMansonQuotes(scrapy.Spider):
    name = "MarilynMansonQuotes"
    start_urls = [
        'https://en.wikiquote.org/wiki/Marilyn_Manson',
    ]
    
    def parse(self, response):
        for quote in response.css('div.mw-parser-output > ul > li'):
            yield {'quote': quote.extract()}

process = CrawlerProcess()
process.crawl(MarilynMansonQuotes)
process.start()
```
